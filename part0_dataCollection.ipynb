```python
def clean(data):
    temp = data.dropna(how='all') #drop the empty days
    cleaned = temp.dropna(axis=1)
    return cleaned
```

```python
## getting S&P500 data tickers , commenting these out for consistency (since I already saved the tickers when I ran it first, which was July 21st 2020)
# names =  pd.read_html('https://en.wikipedia.org/w/index.php?title=List_of_S%26P_500_companies&oldid=472082984')
# table = names[0]
# tickerList = table['Ticker symbol'].tolist()

tickerData = pd.read_csv(path+"data/yfinance/tickers8year.csv",header = None)
tickerList = tickerData.T.values.flatten().tolist()
```

## Downloading Data from Yahoo Finance  

```python
## DUE TO STOCK TICKERS CHANGING, YOU WON'T GET THE SAME DATA IF CALLED AT A LATER POINT.
# THIS WAS COLLECTED ON SEPT 11, 2020.
# AS OF JUNE 21, 2023, ONLY 343 STOCKS OF THE 374 STOCKS CAN BE OBTAINED WITH THIS METHOD

dataAll = yf.download(tickerList, start="2012-01-20", end="2020-01-01") #this gets it until 2019-12-31
lastDay = yf.download(tickerList, start="2019-12-31", end="2020-01-03") #This gets dec 31-jan2nd  so I can return for jan2nd- note that jan 1 has no data
```
